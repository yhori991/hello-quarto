```python
# | echo: true
import plotly.express as px
import polars as pl
import io
import PIL.Image

data = pl.DataFrame(
    {
        "x": [1, 2, 3],
        "y": [2, 2**2, 2**8],
    }
)
fig = px.scatter(data_frame=data, x="x", y="y")

# --- render as png ---
buf = io.BytesIO()
fig.write_image(buf, engine="kaleido", scale=10)
img = PIL.Image.open(buf)
img
```
